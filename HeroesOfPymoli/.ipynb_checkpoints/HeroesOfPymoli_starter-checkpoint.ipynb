* Optional: give the displayed data cleaner formatting

```python
female_purchase = purchase_data.loc[purchase_data['Gender'] == 'Female']
purchase_ct_fe = female_purchase['Gender'].count()
avg_prc_fe = female_purchase['Price'].mean()
total_val_fe = female_purchase['Price'].sum()
avg_total_fe = total_val_fe / gender_df[0]

male_purchase = purchase_data.loc[purchase_data['Gender'] == 'Male']
purchase_ct_ml = male_purchase['Gender'].count()
avg_prc_ml = male_purchase['Price'].mean()
total_val_ml = male_purchase['Price'].sum()
avg_total_ml = total_val_ml / gender_df[0]

# OND Calculations
ond_purchase = purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed']
purchase_ct_ond = ond_purchase['Gender'].count()
avg_prc_ond = ond_purchase['Price'].mean()
total_val_ond = ond_purchase['Price'].sum()
avg_total_ond = total_val_ond / gender_df[2]


## Purchase analysis dictionary
purchasing_analysis = {'Gender': ['Female', 'Male', 'Other / Non-Disclosed'],
                      'Purchase Count': [purchase_ct_fe, purchase_ct_ml, purchase_ct_ond],
                      'Average Purchase Price': [avg_prc_fe, avg_prc_ml, avg_prc_ond],
                      'Total Purchase Value': [total_val_fe, total_val_ml, total_val_ond],
                      'Average Total Purchase per Person':[avg_total_fe, avg_total_ml, avg_total_ond]}

# purchase dictionary to dataframe
purchasing_analysis = pd.DataFrame(purchasing_analysis)

# Set the index to the 'Gender' column
purchasing_analysis.set_index('Gender', inplace=True)

# Format my columns containing dollar values
purchasing_analysis['Average Purchase Price'] = purchasing_analysis['Average Purchase Price'].map("${:.2f}".format)
purchasing_analysis['Total Purchase Value'] = purchasing_analysis['Total Purchase Value'].map("${:,.2f}".format)
purchasing_analysis['Average Total Purchase per Person'] = purchasing_analysis['Average Total Purchase per Person'].map("${:,.2f}".format)

purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]

age_labels = [' < 10', '10-14', '15-19', '20-24', '25-29', '30-34', '34-39', '40+']
```

```python
# Resets the Index
gender_prep_df.reset_index(inplace=True)

# Creates the bins table
age_demog = pd.cut(gender_prep_df['Age'], age_bins, labels=age_labels)

# Creates the dataframe from the bins
gender_prep_df["Age Ranges"] = age_demog
```

```python
# Uses groupby to group the various age ranges
age_group = gender_prep_df.groupby('Age Ranges')

# Counts the number of unique rows that fall into the age ranges
demog_df = age_group['Age'].count()

# Converts all into a dataframe
demog_df = pd.DataFrame(demog_df)
demog_df = demog_df.rename(columns={'Age': 'Total Count'})
total_count = demog_df['Total Count'].sum()
demog_df['Percentage of Players'] = (demog_df['Total Count'] * 100) / total_count
demog_df['Percentage of Players'] = demog_df['Percentage of Players'].map("{:.2f}%".format)
demog_df
```
